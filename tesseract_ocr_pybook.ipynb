```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
import os
import pandas
import re
```

```python
path = 'D:\\mydoc\\twstock\\docs\\ch20\\'

files = []
# r=root, d=directories, f=files
for r,d,f in os.walk(path):
    for file in f:
        if '.jpg' in file:
            files.append(os.path.join(r,file))
```

```python
# 定義 regex
ptn1=re.compile(r',')
ptn2=re.compile(r'\(')
ptn3=re.compile(r'\)')
ptn4=re.compile(r':')
ptn5=re.compile(r'\?')
ptn6=re.compile(r' ')


# 產生orc結果
result=''
outfile=open(r"D:\\mydoc\\twstock\\docs\\ch20.txt","a")
for f in files:
    res=pytesseract.image_to_string(Image.open(f), lang='chi_tra',config='--psm 6')
    res=ptn6.sub('',ptn5.sub('？',ptn4.sub('：',ptn3.sub('）',ptn2.sub('（',ptn1.sub('，',res))))))
    outfile.write(res)

outfile.close()
```

```python
# 導入庫
import numpy as np
import argparse
import cv2

# 定義平移translate函數
def translate(image, x, y):
    # 定義平移矩陣
    M = np.float32([[1, 0, x], [0, 1, y]])
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
 
    # 返回轉換後的圖像
    return shifted


# 構造參數解析器
ap = argparse.ArgumentParser()
ap.add_argument("-i", "--image", required=True, help="Path to the image")
args = vars(ap.parse_args())

# 加載圖像並顯示
image = cv2.imread(args["image"])
cv2.imshow("Original", image)

# 將原圖分別做上、下、左、右平移操作
shifted = translate(image, 0, 100)
cv2.imshow("Shifted Down", shifted)
shifted = translate(image, 0, -100)
cv2.imshow("Shifted Up", shifted)
shifted = translate(image, 50, 0)
cv2.imshow("Shifted Right", shifted)
shifted = translate(image, -50, 0)
cv2.imshow("Shifted Left", shifted)
cv2.waitKey(0)
```

```python
# 導入庫
import numpy as np
import argparse
import cv2

# 定義旋轉rotate函數
def rotate(image, angle, center=None, scale=1.0):
    # 獲取圖像尺寸
    (h, w) = image.shape[:2]
 
    # 若未指定旋轉中心，則將圖像中心設為旋轉中心
    if center is None:
        center = (w / 2, h / 2)
 
    # 執行旋轉
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
 
    # 返迴旋轉後的圖像
    return rotated

# 構造參數解析器
ap = argparse.ArgumentParser()
ap.add_argument("-i", "--image", required=True, help="Path to the image")
args = vars(ap.parse_args())
 
# 加載圖像並顯示
image = cv2.imread(args["image"])
cv2.imshow("Original", image)
 
# 將原圖旋轉不同角度
rotated = rotate(image, 45)
cv2.imshow("Rotated by 45 Degrees", rotated)
rotated = rotate(image, -45)
cv2.imshow("Rotated by -45 Degrees", rotated)
rotated = rotate(image, 90)
cv2.imshow("Rotated by 90 Degrees", rotated)
rotated = rotate(image, -90)
cv2.imshow("Rotated by -90 Degrees", rotated)
rotated = rotate(image, 180)
cv2.imshow("Rotated by 180 Degrees", rotated)
cv2.waitKey(0)
```

```python
# 導入庫
import argparse
import cv2
 
# 定義縮放resize函數
def resize(image, width=None, height=None, inter=cv2.INTER_AREA):
    # 初始化縮放比例，並獲取圖像尺寸
    dim = None
    (h, w) = image.shape[:2]
 
    # 如果寬度和高度均為0，則返回原圖
    if width is None and height is None:
        return image
 
    # 寬度是0
    if width is None:
        # 則根據高度計算縮放比例
        r = height / float(h)
        dim = (int(w * r), height)
 
    # 如果高度為0
    else:
        # 根據寬度計算縮放比例
        r = width / float(w)
        dim = (width, int(h * r))
 
    # 縮放圖像
    resized = cv2.resize(image, dim, interpolation=inter)
 
    # 返回縮放後的圖像
    return resized


# 構造參數解析器
ap = argparse.ArgumentParser()
ap.add_argument("-i", "--image", required=True, help="Path to the image")
args = vars(ap.parse_args())
 
# 加載參數並顯示（由於原圖較大，我們先將原圖縮小）
image = cv2.imread(args["image"])
cv2.imshow("Original", image)

# 創建插值方法數組
methods = [
    ("cv2.INTER_NEAREST", cv2.INTER_NEAREST),
    ("cv2.INTER_LINEAR", cv2.INTER_LINEAR),
    ("cv2.INTER_AREA", cv2.INTER_AREA),
    ("cv2.INTER_CUBIC", cv2.INTER_CUBIC),
    ("cv2.INTER_LANCZOS4", cv2.INTER_LANCZOS4)]
 
# 循環執行插值方法數組中的各個方法
for (name, method) in methods:
    # 放大3倍
    resized = resize(image, width=image.shape[1] * 3, inter=method)
    cv2.imshow("ZoomIn Method: {}".format(name), resized)
    # 縮小2倍
    resized = resize(image, width=image.shape[1] /2, inter=method)
    cv2.imshow("ZoomOut Method: {}".format(name), resized)
    cv2.waitKey(0)
```

```python
# 導入庫
import argparse
import cv2
 
# 構造參數解析器
ap = argparse.ArgumentParser()
ap.add_argument("-i", "--image", required=True, help = "Path to the image")
args = vars(ap.parse_args())
 
# 加載圖像並顯示
image = cv2.imread(args["image"])
cv2.imshow("Original", image)
 
# 橫向翻轉圖像
flipped = cv2.flip(image, 1)
cv2.imshow("Flipped Horizontally", flipped)
 
# 縱向翻轉圖像
flipped = cv2.flip(image, 0)
cv2.imshow("Flipped Vertically", flipped)
 
# 同時在橫向和縱向翻轉圖像
flipped = cv2.flip(image, -1)
cv2.imshow("Flipped Horizontally & Vertically", flipped)
cv2.waitKey(0)
```
